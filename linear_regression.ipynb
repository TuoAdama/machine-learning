```python
def normalization(dataToNormalize):
    columns = dataToNormalize.columns
    for column in columns:
        x = dataToNormalize[column].values.astype(float).reshape(-1, 1)
        standard_normalization = preprocessing.StandardScaler()
        res = standard_normalization.fit_transform(x)
        dataToNormalize[column] = res.flatten()
```

#### train_index contient les indices parmi ceux présents dans le data-set et test_index contient les indices qui ne sont pas dans train_index.

```python
def create_evaluate_model(index_fold, x_train, x_test, y_train, y_test):
    regression_algo = LinearRegression()
    regression_algo.fit(x_train, y_train)
    tests_predictions = regression_algo.predict(x_test)
    rmse = sqrt(mean_squared_error(y_test, tests_predictions))
    r2 = r2_score(y_test, tests_predictions)
    print(f"Fold {index_fold} : RMSE = {round(rmse, 2)} - R2 = {round(r2, 2)}")
    return (rmse, r2)
```

La RMSE (Root Mean Square Error, ou racine de l’erreur quadratique moyenne) est une mesure de performance d’un modèle de machine learning en régression.
Elle permet d’évaluer à quel point les prédictions du modèle sont éloignées des valeurs réelles.

$$
RMSE = \sqrt{\frac{1}{n} \sum (y_{réel} - y_{prédit})^2}
$$

	•	RMSE = 0 → modèle parfait
	•	RMSE faible → bonnes prédictions
	•	RMSE élevée → modèle imprécis

Le R² score (ou coefficient de détermination) est une autre métrique clé pour évaluer un modèle de régression.

$$
R^2 = 1 - \frac{\sum (y_i - \hat{y}_i)^2}{\sum (y_i - \bar{y})^2}
$$


	•	R² = 1 → modèle parfait
	•	R² = 0 → aussi bon que prédire la moyenne
	•	R² < 0 → pire que prédire la moyenne
	•	R² ≈ 0.8 → le modèle explique 80 % de la variance des données

```python
split = 5
kf = KFold(n_splits=split, shuffle=False)
index_fold = 0
average_rmse = 0
average_r2 = 0

for train_index, test_index in kf.split(data):
    x_train, x_test = data.iloc[train_index], data.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]
    rmse, r2 = create_evaluate_model(index_fold, x_train, x_test, y_train, y_test)
    average_rmse += rmse
    average_r2 += r2
    index_fold += 1

average_rmse /= split
average_r2 /= split

print(f"Average RMSE = {round(average_rmse, 2)} - Average R2 = {round(average_r2, 2)}")
```
